A W-state circuit generate 3 entangled qubits 

Step:1 Initialise a 3 qubit circuit
Step:2 perform a Ry rotation on the qubit (provided)

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import plot_histogram
```

```python
qc = QuantumCircuit(3)
qc.ry(1.91063324, 0)
```

# Steps to build the W-state circuit
### Perform a controlled hadamard gate on qubit 1, with control qubit 0
### Add a CNOT gate with control qubit 1 and target qubit 2
### Add a CNOT gate with control qubit 0 and target qubit 1
### Add a X gate on qubit 0

```python
#Controlled hadmard gate with control quibt, target qubit
qc.ch(0, 1)
qc.cx(1, 2)
qc.cx(0, 1)
qc.x(0)
qc.measure_all()
qc.draw('mpl')
```

```python
sampler = StatevectorSampler()
pub = (qc)
job_sampler = sampler.run([pub], shots=10000)

result_sampler = job_sampler.result()
counts_sampler = result_sampler[0].data.meas.get_counts()

print(counts_sampler)
```

```python
plot_histogram(counts_sampler)
```
